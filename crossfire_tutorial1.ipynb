### Importing modules

````python
import crossfire
from geobr import read_municipality
import matplotlib.pyplot as plt
````

### Downloading data

````python
rj = read_municipality(33)
````

**Before downloading crossfire data, [you must have a login](https://fogocruzado.org.br/sobre-a-api/)**.  

And then, **signin, as follow**:

````python
# crossfire.fogocruzado_signin('user@host.com', 'password')
````

**Then you can retrieve data with `get_fogocruzado`**

````python
tiroteios = crossfire.get_fogocruzado(state='RJ')

tiroteios.head()
````

It already comes as GeoDataFrame class:

````python
type(tiroteios)
````

### Ploting all data

````python
tiroteios.plot()
````

Let's plot together with municipalities bounds downloaded from [`GeoBr`](https://pypi.org/project/geobr/) project.

````python
f, ax = plt.subplots(figsize=(15,15))
rj.plot(ax = ax)
tiroteios.plot(ax=ax, color='k')
````
