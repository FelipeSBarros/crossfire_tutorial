### Importing modules

````python
import crossfire
from geobr import read_municipality
import matplotlib.pyplot as plt
````

### Downloading data

````python
rj = read_municipality(33)
````

**Before downloading crossfire data, you must signin:**

````python
# crossfire.fogocruzado_signin('user@host.com', 'password')
````

**Then you can retrieve data with `get_fogocruzado`**

````python
tiroteios = crossfire.get_fogocruzado(state='RJ')

tiroteios.head()
````

````python
type(tiroteios)
````

````python
tiroteios.plot()
````

````python
f, ax = plt.subplots(figsize=(15,15))
rj.plot(ax = ax)
tiroteios.plot(ax=ax, color='k')
````
