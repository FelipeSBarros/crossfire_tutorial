### Importing modules

````python
import crossfire
from geobr import read_municipality
import geoplot as gplt
````

### Downloading data

````python
rj = read_municipality(33)
````

**Before downloading crossfire data, [you must have a login](https://fogocruzado.org.br/sobre-a-api/)**.  

And then, **signin, as follow**:

````python
# crossfire.fogocruzado_signin('user@host.com', 'password')
````

**Then you can retrieve data with `get_fogocruzado`**

````python
tiroteios = crossfire.get_fogocruzado(state='RJ')
````

### Ploting fogo cruzado records with `kdeplot`

Plotting with [`Kddeplot`](https://residentmario.github.io/geoplot/plot_references/plot_reference.html#kdeplot)

````python
ax = gplt.polyplot(rj)
gplt.kdeplot(tiroteios, ax=ax)
````

Using kdeplot with `shade=True`

````python
ax = gplt.polyplot(rj)
gplt.kdeplot(tiroteios, ax=ax, shade=True)
````

Ploting the KDE as shade and cliped by municipalities bounds:

````python
ax = gplt.polyplot(rj)
gplt.kdeplot(tiroteios, ax=ax, shade=True, clip=rj.geometry)
````
